```python
from sklearn.preprocessing import LabelEncoder

# Reconstruct df_new from the original df and drop 'Loan_ID'
df_new = df.drop(['Loan_ID'], axis=1)

# Re-apply Label Encoding for 'Education'
le_education = LabelEncoder()
df_new['Education'] = le_education.fit_transform(df_new['Education'])

# Re-apply Label Encoding for 'Gender'
le_gender = LabelEncoder()
df_new['Gender'] = le_gender.fit_transform(df_new['Gender'])

# Re-apply Label Encoding for 'Married'
le_married = LabelEncoder()
df_new['Married'] = le_married.fit_transform(df_new['Married'])

# Re-apply Label Encoding for 'Self_Employed'
le_self_employed = LabelEncoder()
df_new['Self_Employed'] = le_self_employed.fit_transform(df_new['Self_Employed'])

# Now, apply one-hot encoding for 'Property_Area'
df_new = pd.get_dummies(df_new, columns=["Property_Area"], prefix="Property_Area", drop_first=True)
df_new.head()
```

```python
bool_cols=df_new.select_dtypes(include='bool').columns
df_new[bool_cols]=df_new[bool_cols].astype(int)
```

```python
y=df['Loan_Status']
x=df.drop(['Loan_Status'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
y_predict=lr_model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_predict)
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier() #without feature
rf_model.fit(x_train,y_train)
y_predict=rf_model.predict(x_test)
print(accuracy_score(y_test,y_predict))
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
y_pred=dt_model.predict(x_test)
print(accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```
